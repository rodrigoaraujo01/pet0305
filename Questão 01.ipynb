# Solução da Questão 01

a) Sigmóide: $$\varphi(v) = \frac{1}{1+exp(-av)}$$

b) Tangente sigmóide: $$\varphi(v) = \frac{1-exp(-av)}{1+exp(-av)}$$

c) ReL (retificador linear): $$\varphi(v) = max(0,v)$$

## i) Plot das funções

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
sigmoide = lambda a, x: 1/(1+np.exp(-a*x))
plt.plot(x, sigmoide(1, x))
plt.title('Sigmóide')
plt.show()
```

```python
tg_sigmoide = lambda a, x: (1-np.exp(-a*x))/(1+np.exp(-a*x))
plt.plot(x, tg_sigmoide(1, x))
plt.title('Tangente sigmóide')
plt.show()
```

```python
rel = lambda x: max(0, x)
y = [rel(i) for i in x]
plt.plot(x, y)
plt.title('Retificador linear')
plt.show()
```

```python
plt.plot(x, sigmoide(1, x), label='sigmóide')
plt.plot(x, tg_sigmoide(1, x), label='tangente sigmóide')
plt.plot(x, y, label='retificador linear')
plt.title('Comparação')
plt.legend()
plt.show()
```

## ii) Derivada da função sigmóide

$$\frac{d\varphi}{dv} = \frac{(1)'\cdot(1+exp(-av)) - (1)\cdot(1+exp(-av))'}{(1+exp(-av))^2}=\frac{-(-a\cdot exp(-av))}{(1+exp(-av))^2}$$

$$\frac{d\varphi}{dv} = a \cdot \varphi(v) \cdot \frac{exp(-av)}{(1+exp(-av))}$$

$$1-\varphi(v) = 1 - \frac{1}{1+exp(-av)} = \frac{1+exp(-av)-1}{1+exp(-av)} = \frac{exp(-av)}{1+exp(-av)}$$

$$\frac{d\varphi}{dv} = a \cdot \varphi(v) \cdot [1-\varphi(v)]$$

## iii) Derivada da função tangente sigmóide

b) Tangente sigmóide: $$\varphi(v) = \frac{1-exp(-av)}{1+exp(-av)}$$

$$\frac{d\varphi}{dv} = \frac{(1-exp(-av))'\cdot(1+exp(-av)) - (1-exp(-av))\cdot(1+exp(-av))'}{(1+exp(-av))^2}$$

$$\frac{d\varphi}{dv} =\frac{(-exp(-av))\cdot(-a)\cdot(1+exp(-av))-(1-exp(-av)) \cdot exp(-av)\cdot (-a)}{(1+exp(-av))^2}$$

$$\frac{d\varphi}{dv} =\frac{a\cdot(exp(-av))\cdot(1+exp(-av))+a\cdot(exp(-av))\cdot(1-exp(-av))}{(1+exp(-av))^2}$$

$$\frac{d\varphi}{dv}=\frac{a \cdot exp(-av)\cdot(1+exp(-av)+1-exp(-av))}{(1+exp(-av))^2}$$

$$\frac{d\varphi}{dv}=\frac{a \cdot exp(-av)\cdot2}{(1+exp(-av))^2}$$

$$\frac{d\varphi(v)}{dv}=\frac{a}{2}\cdot\frac{4\cdot exp(-av)}{(1+exp(-av))^2}$$

$$1-\varphi^2(v) = 1 - \frac{(1-exp(-av))^2}{(1+exp(-av))^2}$$

$$1-\varphi^2(v) = \frac{(1+exp(-av))^2-(1-exp(-av))^2}{(1+exp(-av))^2}$$

$$1-\varphi^2(v) = \frac{(1+exp(-av)+1-exp(-av))\cdot(1+exp(-av)-(1-exp(-av))}{(1+exp(-av))^2}=\frac{4\cdot exp(-av)}{(1+exp(-av))^2}$$

$$\frac{d\varphi}{dv} = \frac{a}{2} \cdot [1-\varphi^2(v)]$$

## iv) Análise com $$v = w^t x = x^t w$$

a) Sigmóide
$$\frac{d\varphi(v)}{dv} = a \cdot \varphi(v) \cdot [1-\varphi(v)]$$

Sabe-se que

$$\frac{d\textbf{u}^t\textbf{x}}{d\textbf{x}}=\textbf{u}^t$$

Logo

$$\frac{d\varphi(\textbf{x}^t \textbf{w})}{d\textbf{w}} = \frac{d\varphi(\textbf{x}^t \textbf{w})}{d\textbf{x}^t \textbf{w}} \cdot \frac{d\textbf{x}^t\textbf{w}}{d\textbf{w}}$$

$$\frac{d\varphi(\textbf{x}^t \textbf{w})}{d\textbf{x}^t \textbf{w}} = a \cdot \varphi(\textbf{x}^t \textbf{w}) \cdot [1 - \varphi(\textbf{x}^t \textbf{w})]$$

$$\frac{d\textbf{x}^t\textbf{w}}{d\textbf{w}} = \textbf{x}^t$$

$$\frac{d\varphi(\textbf{x}^t \textbf{w})}{d\textbf{w}} = a \cdot \varphi(\textbf{x}^t \textbf{w}) \cdot [1 - \varphi(\textbf{x}^t \textbf{w})] \cdot \textbf{x}^t$$

b) Tangente sigmóide
$$\frac{d\varphi}{dv} = \frac{a}{2} \cdot [1-\varphi^2(v)]$$

Sabe-se que

$$\frac{d\textbf{u}^t\textbf{x}}{d\textbf{x}}=\textbf{u}^t$$

Logo

$$\frac{d\varphi(\textbf{x}^t \textbf{w})}{d\textbf{w}} = \frac{d\varphi(\textbf{x}^t \textbf{w})}{d\textbf{x}^t \textbf{w}} \cdot \frac{d\textbf{x}^t\textbf{w}}{d\textbf{w}}$$

$$\frac{d\varphi(\textbf{x}^t \textbf{w})}{d\textbf{x}^t \textbf{w}} = \frac{a}{2} \cdot [1 - \varphi^2(\textbf{x}^t \textbf{w})]$$

$$\frac{d\textbf{x}^t\textbf{w}}{d\textbf{w}} = \textbf{x}^t$$

$$\frac{d\varphi(\textbf{x}^t \textbf{w})}{d\textbf{w}} = \frac{a}{2} \cdot [1 - \varphi^2(\textbf{x}^t \textbf{w})] \cdot \textbf{x}^t$$
